```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
file_path = 'TILL6022_Emission_Dataset.csv'
df = pd.read_csv(file_path, delimiter=',')
df
```

```python
# filter dataset on Transport sectors: 'International Shipping' or 'International Aviation' or 'Domestic Aviation' or 'Ground Transport'
sectors = ['International Shipping', 'International Aviation', 'Domestic Aviation', 'Ground Transport']
Transport_Sector =  df[df.sector.isin(sectors)]
print(Transport_Sector.head())


```

```python
# exclude 'WORLD' from dataframe

Transport_not_World = Transport_Sector[Transport_Sector.country != 'WORLD'].groupby(['country', 'date']).sum()

#reset index for optimal dataset and split date into years

abc = Transport_not_World.reset_index(['country', 'date'])
abc[["year", "month", "day"]] = abc["date"].str.split("-", expand = True)



print(abc)

df_UK_2019 =  abc[(abc.country == 'UK') & (abc.year == '2019')] 
df_UK_2020 =  abc[(abc.country == 'UK') & (abc.year == '2020')]
df_UK_2020['Date2'] = '2019' + '-' + df_UK_2020['month'].astype(str) + '-' + df_UK_2020['day']

df_USA_2019 =  abc[(abc.country == 'US') & (abc.year == '2019')]
df_USA_2020 =  abc[(abc.country == 'US') & (abc.year == '2020')]
df_USA_2020['Date2'] = '2019' + '-' + df_USA_2020['month'].astype(str) + '-' + df_USA_2020['day']

df_CHINA_2019 = abc[(abc.country == 'China') & (abc.year == '2019')]
df_CHINA_2020 = abc[(abc.country == 'China') & (abc.year == '2020')]
df_CHINA_2020['Date2'] = '2019' + '-' + df_CHINA_2020['month'].astype(str) + '-' + df_CHINA_2020['day']
# create figure
fig = go.Figure()

# add trace
fig.add_trace(
    go.Scatter(x=list(df_UK_2019.date),
               y=list(df_UK_2019.co2),
               name="UK 2019",
               line=dict(color="#33CFA5")))

fig.add_trace(
    go.Scatter(x=list(df_UK_2020.Date2),
               y=list(df_UK_2020.co2),
               name="UK 2020",
               line=dict(color="#c2fc03")))

fig.add_trace(
    go.Scatter(x=list(df_USA_2019.date),
               y=list(df_USA_2019.co2),
               name="USA 2019",
               line=dict(color="#F06A6A")))

fig.add_trace(
    go.Scatter(x=list(df_USA_2020.Date2),
               y=list(df_USA_2020.co2),
               name="USA 2020",
               line=dict(color="#fc0390")))

fig.add_trace(
    go.Scatter(x=list(df_CHINA_2019.date),
               y=list(df_CHINA_2019.co2),
               name="China 2019",
               line=dict(color="#32CD32")))

fig.add_trace(
    go.Scatter(x=list(df_CHINA_2020.Date2),
               y=list(df_CHINA_2020.co2),
               name="China 2020",
               line=dict(color="#5e03fc")))

fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="UK",
                     method="update",
                     args=[{"visible": [True, True, False, False, False, False]}]),
                dict(label="USA",
                     method="update",
                     args=[{"visible": [False, False, True, True, False, False]}]),
                dict(label="China",
                     method="update",
                     args=[{"visible": [False, False, False, False, True, True]}])
                     

]))])

fig.update_xaxes(
    dtick="M1",
    tickformat="%b")
                        
fig.show()
```

```python
# exclude 'WORLD' from dataframe

Transport_not_World = Transport_Sector[Transport_Sector.country != 'WORLD'].groupby(['country', 'date']).sum()


#reset index for optimal dataset and split date into years

abc = Transport_not_World.reset_index(['country', 'date'])
abc[["year", "month", "day"]] = abc["date"].str.split("-", expand = True)

# create dataframe for sector

Transport_Sector[["year", "month", "day"]] = Transport_Sector["date"].str.split("-", expand = True)
print(Transport_Sector)

df_UK_2019_Ground = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2019') & 
                                    (Transport_Sector.sector == 'Ground Transport')]
df_UK_2019_Ship = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2019') & 
                                    (Transport_Sector.sector == 'International Shipping')]
df_UK_2019_Int_Av = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2019') & 
                                    (Transport_Sector.sector == 'International Aviation')]
df_UK_2019_Do_Av = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2019') & 
                                    (Transport_Sector.sector == 'Domestic Aviation')]
df_UK_2019_Total =  abc[(abc.country == 'UK') & (abc.year == '2019')] 

# for the 2020 data, we need to align it with 2019 data

Transport_Sector['Date2'] = '2019' + '-' + Transport_Sector['month'].astype(str) + '-' + Transport_Sector['day']

# create a unique dataset for each sector to easily add trace

df_UK_2020_Ground = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2020') & 
                                    (Transport_Sector.sector == 'Ground Transport')]
df_UK_2020_Ship = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2020') & 
                                    (Transport_Sector.sector == 'International Shipping')]
df_UK_2020_Int_Av = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2020') & 
                                    (Transport_Sector.sector == 'International Aviation')]
df_UK_2020_Do_Av = Transport_Sector[(Transport_Sector.country == 'UK') & (Transport_Sector.year == '2019') & 
                                    (Transport_Sector.sector == 'Domestic Aviation')]
df_UK_2020_Total =  abc[(abc.country == 'UK') & (abc.year == '2020')]
df_UK_2020_Total['Date2'] = '2019' + '-' + df_UK_2020_Total['month'].astype(str) + '-' + df_UK_2020_Total['day']


df_USA_2019 =  abc[(abc.country == 'US') & (abc.year == '2019')]
df_USA_2020 =  abc[(abc.country == 'US') & (abc.year == '2020')]
df_USA_2020['Date2'] = '2019' + '-' + df_USA_2020['month'].astype(str) + '-' + df_USA_2020['day']

df_CHINA_2019 = abc[(abc.country == 'China') & (abc.year == '2019')]
df_CHINA_2020 = abc[(abc.country == 'China') & (abc.year == '2020')]
df_CHINA_2020['Date2'] = '2019' + '-' + df_CHINA_2020['month'].astype(str) + '-' + df_CHINA_2020['day']
# create figure
fig = go.Figure()

# add trace
fig.add_trace(
    go.Scatter(x=list(df_UK_2019_Total.date),
               y=list(df_UK_2019_Total.co2),
               name="UK 2019",
               visible=False,
               line=dict(color="#33CFA5")))

fig.add_trace(
    go.Scatter(x=list(df_UK_2020_Total.Date2),
               y=list(df_UK_2020_Total.co2),
               name="UK 2020",
               visible =False,
               line=dict(color="#c2fc03")))

fig.add_trace(
    go.Scatter(x=list(df_UK_2019_Ground.date),
               y=list(df_UK_2019_Ground.co2),
               name="UK 2019 Ground Transport",
               visible=False,
               line=dict(color="#33CFA5")))

fig.add_trace(
    go.Scatter(x=list(df_UK_2020_Ground.Date2),
               y=list(df_UK_2020_Ground.co2),
               name="UK 2020 Ground Transport",
               visible =False,
               line=dict(color="#c2fc03")))

fig.add_trace(
    go.Scatter(x=list(df_USA_2019.date),
               y=list(df_USA_2019.co2),
               name="USA 2019",
               visible=False,
               line=dict(color="#F06A6A")))

fig.add_trace(
    go.Scatter(x=list(df_USA_2020.Date2),
               y=list(df_USA_2020.co2),
               name="USA 2020",
               visible = False,
               line=dict(color="#fc0390")))

fig.add_trace(
    go.Scatter(x=list(df_CHINA_2019.date),
               y=list(df_CHINA_2019.co2),
               name="China 2019",
               visible= False,
               line=dict(color="#32CD32")))

fig.add_trace(
    go.Scatter(x=list(df_CHINA_2020.Date2),
               y=list(df_CHINA_2020.co2),
               name="China 2020",
               visible= False,
               line=dict(color="#5e03fc")))








fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(label="UK",
                     method="restyle",
                     args=[{"visible": [True, True, False, False, False, False, False, False]}]),
                dict(label="USA",
                     method="restyle",
                     args=[{"visible": [False, False, False, False, True, True, False, False]}]),
                dict(label="China",
                     method="restyle",
                     args=[{"visible": [False, False,False, False, False, False, True, True]}])
                    ]),

        direction = 'down',
        pad = {'r': 10, 't': 10},
        showactive = True,
        x = 0.1,
        xanchor = 'left',
        y = 1.5,
        yanchor = 'top'
    ),
        dict(
            buttons=list([
                dict(label="Ground",
                     method="restyle",
                     args=[{"visible": [False, False, True, True, False, False, False, False]}])
            ]),
        direction = 'down',
        pad = {'r': 10, 't': 10},
        showactive = True,
        x = 0.55,
        xanchor = 'left',
        y = 1.5,
        yanchor = 'top'
        ),    
])

fig.update_xaxes(
    dtick="M1",
    tickformat="%b")
                        
fig.show()
```
